# 4. Create a new column in the dataframe “age_group”. This column should contain the age group the respondent belongs to. The age groups are 18-25, 25-40, 40-55 and 55+. 

```python
dfd = survey.copy()
dfd.insert(10, 'age_group', survey.age)  #insert(loc, 'column name', value)

dfd.age_group.replace({'24ko':'24'}, inplace=True)
dfd.age_group.unique()
dfd.age_group = dfd.age_group.astype(int)
dfd.age_group.unique()
```

```python
data1 = dfd[(dfd.age_group >=18) & (dfd.age_group <=25)]
data1['age_group'] = '18-25'
data1

data2 = dfd[(dfd.age_group >= 25) & (dfd.age_group < 40)]
data2['age_group'] = '25-40'
data2

data3 = dfd[(dfd.age_group >= 40) & (dfd.age_group < 55)]
data3['age_group'] = '40-55'
data3

data4 = dfd[dfd.age_group >= 55]
data4['age_group'] = '55+'
data4

data = pd.concat([data1, data2, data3, data4])
data['age_group'].unique()

data
```
